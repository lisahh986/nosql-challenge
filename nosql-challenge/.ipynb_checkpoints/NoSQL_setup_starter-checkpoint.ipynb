## Part 2: Update the Database

1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
penang_flavours_restaurant = {
                              "BusinessName": "Penang Flavours",
                              "BusinessType": "Restaurant/Cafe/Canteen",
                              "BusinessTypeID": "",
                              "AddressLine1": "Penang Flavours",
                              "AddressLine2": "146A Plumstead Rd",
                              "AddressLine3": "London",
                              "AddressLine4": "",
                              "PostCode": "SE18 7DY",
                              "Phone": "",
                              "LocalAuthorityCode": "511",
                              "LocalAuthorityName": "Greenwich",
                              "LocalAuthorityWebSite": "http://www.royalgreenwich.gov.uk",
                              "LocalAuthorityEmailAddress": "health@royalgreenwich.go.uk",
                              "scores": {
                                "Hygiene": "",
                                "Structural": "",
                                "ConfidenceInManagement": ""
                              },
                              "SchemeType": "FHRS",
                              "geocode": {
                                "longitude": "0.08384000",
                                "latitude": "51.49014200"
                              },
                              "RightToReply": "",
                              "Distance": 4623.972328074718,
                              "NewRatingPending":True
                            }
```

```python
# Insert the new restaurant into the collection
establishmentsCollection.insert_one(penang_flavours_restaurant)
```

```python
# Check that the new restaurant was inserted
query_01 = {'BusinessName': 'Penang Flavours'}
# pprint(establishmentsCollection.find_one(query_01))
results_01 = establishmentsCollection.find(query_01)
for result in results_01:
    pprint(result)
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
# results_02 = {'BusinessTypeID': "Restaurant/Cafe/Canteen"}
# query_02 = {'BusinessTypeID': "Restaurant/Cafe/Canteen"}
# fields_02 = {"BusinessTypeID", "BusinessType"}
# results_02 = list(establishmentsCollection.find(query_02, fields_02))
# for result in results_02:
#     pprint(result)
    
query_02 = {'BusinessType': 'Restaurant/Cafe/Canteen'}
fields_02 = ['BusinessTypeID', 'BusinessType']
result_03= establishmentsCollection.find_one(query_02, fields_02)
id = result_03['BusinessTypeID']
print(f'Founded document: {result_03}')
print(f'BusinessTypeID: {id}')
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
establishmentsCollection.update_one(
    penang_flavours_restaurant, 
    {'$set': 
        {'BusinessTypeID': 1}
    }
)
```

```python
# Confirm that the new restaurant was updated
pprint(establishmentsCollection.find_one( {'BusinessName': 'Penang Flavours'} ) )
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
numOfDover = establishmentsCollection.count_documents( {"LocalAuthorityName":"Dover"} )
print(f'How many documents have LocalAuthorityName as "Dover": {numOfDover}')
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
establishmentsCollection.delete_many( {"LocalAuthorityName":"Dover"} )
```

```python
# Check if any remaining documents include Dover
establishmentsCollection.count_documents( {"LocalAuthorityName":"Dover"} )
```

```python
# Check that other documents remain with 'find_one'
establishmentsCollection.find_one({})
```

5. Some of the number values are stored as strings, when they should be stored as numbers. Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude
establishmentsCollection.update_many({}, [{'$set': {'geocode.longitude': {'$toDouble': '$geocode.longitude'}}}])
```

```python
# Change the data type from String to Decimal for latitude
establishmentsCollection.update_many({}, [{'$set': {'geocode.latitude': {'$toDouble': '$geocode.latitude'}}}])
```

```python
# Check that the coordinates are now numbers
query_03 = {}
fields_03 = {"geocode.longitude", "geocode.latitude"}
results_03 = list(establishmentsCollection.find(query_03, fields_03))
for result in results_03:
    pprint(result)
```
